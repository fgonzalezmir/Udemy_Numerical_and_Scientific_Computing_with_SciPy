### Gradient Descent: theoretical background

[The standard (batch) gradient descent](https://en.wikipedia.org/wiki/Gradient_descent) is an optimization algorithm devised to find the values of the parameters $\theta_0$, $\theta_1$, $\dots$, $\theta_n$ for a linear regression problem posed in the form:

\begin{aligned}
        {\textbf h}(\mathbf{\theta}) = \theta_0 {\textbf X}_0 + \theta_1 {\textbf X}_1 + \theta_2 {\textbf X}_2 + \cdots + \theta_n {\textbf X}_n =       
%        
\begin{pmatrix} {\textbf X}_0 \; {\textbf X}_1 \; {\textbf X}_2 \; \cdots \; {\textbf X}_n \end{pmatrix}
        \begin{pmatrix}\theta_0 \\
                       \theta_1 \\
                       \theta_2 \\
                       \vdots \\
                       \theta_n \end{pmatrix}
        %
        \end{aligned}
        
The ${\textbf X}_i$ ($i=1 \cdots n$) are variables or features vectors, containing $m$ observations. ${\textbf X}_0$ is conveniently defined with all its components having the value one. 

### The squared error cost function 

To find the best parameters $\theta_i$, the opimization problem is usually posed in terms of ** *minimizing* ** the (convex) error measure:

\begin{aligned}
        J(\theta) &= \frac{1}{2m}\sum_{i=1}^m \left[ h^{(i)}(\theta)-y^{(i)}\right]^2 \\
                  &= \frac{1}{2m}\sum_{i=1}^m \left[ \theta_0 + \theta_1 X_1^{(i)} + \theta_2 X_2^{(i)} + \cdots + \theta_n X_n^{(i)} - y^{(i)}\right]^2 \\
                  &= \frac{1}{2m}\sum_{i=1}^m \left[ \theta_0 X_0^{(i)} + \theta_1 X_1^{(i)} + \theta_2 X_2^{(i)} + \cdots + \theta_n X_n^{(i)} - y^{(i)}\right]^2,
        \end{aligned}

where the upper index $(i)$ represents the $i_{\text{th}}$ observed (measured) value of the respective variable (here we are assuming there are $m$ observations of the quantity $y$ corresponding to $m$ simultaneuos observations for each one of the $n$ variables on which the quantity $y$ depends on). Here we have added the extra term $X_0$ whose values are all ones.

### The Gradient Descent algorithm

Now we are ready to state the procedure throught which $J(\theta)$ can be minimized according to the ** *gradient descent* ** algorithm. The method requires starting with a set of initial values for the parameters $\theta_0$, $\theta_1$, $\theta_2$,$\cdots$, $\theta_n$ and then
update them according to the rule:

\begin{aligned}
\theta_k = \theta_k - \alpha \frac{\partial J(\theta)}{\!\!\!\! \partial \theta_k}, 
\end{aligned}

where $\alpha$ is a parameter defined by the user called the ** *learning rate.* ** Since $J(\theta)$ is a convex function, it is guarantee it has only a global minimun. The convergence to that minimun following the ** *gradient descent* **  rule could be slow (or never happen) depending on the value given to $\alpha$. Thus there is an optimal value for $\alpha$, but there is no a simple criterium to choose it. s 

The aforementioned update rule for the $\theta$'s is applied as many times (iterations)
as required to reach a reasonable value for the minimun of $J(\theta)$, which is defined by a ** *tolerance* ** value. Accordingly, a rule of thumb to check that the ** *gradient descent* **  rule is working is to plot $J(\theta)$ Vs the number of iterations that we have applied to update the $\theta$'s. We should see a decreasing curve which eventually (after many iterations) should become flat.

In this sense, a general advice to choose the value of $\alpha$ is to try a range of values for it and pick the one that gives fast convergence. One can start with $\alpha = 0.0001$, then $\alpha=0.001$, continuing with $\alpha=0.01$ and so on, according to the behaviour of $J(\theta)$ Vs the number of iteration as already mentioned (if $J(\theta)$ Vs the number of iterations increases is because $\alpha$ is too large if no bug is in your code).

### Derivatives of the squared error cost function

** *Minimizing* ** $J(\theta)$ via the ** *gradient descent* ** algorithm requires the derivative of $J(\theta)$ respect each one of the $\theta_i$'s, which is readily obtained in the form:

\begin{aligned}
        \frac{\partial J(\theta)}{\!\!\!\! \partial \theta_k} 
                  &= \frac{1}{m}\sum_{i=1}^m \left[ \theta_0 X_0^{(i)} + \theta_1 X_1^{(i)} + \theta_2 X_2^{(i)} + \cdots + \theta_n X_n^{(i)} - y^{(i)}\right] X_k^{(i)} \\
                  &= \frac{\theta_0}{m}\sum_{i=1}^m X_0^{(i)} X_k^{(i)} +
                     \frac{\theta_1}{m}\sum_{i=1}^m X_1^{(i)} X_k^{(i)} +
                     \frac{\theta_2}{m}\sum_{i=1}^m X_2^{(i)} X_k^{(i)} + \cdots + \\
                   & \quad\quad\quad\quad\quad\quad  
                   \frac{\theta_n}{m}\sum_{i=1}^m X_n^{(i)} X_k^{(i)} -
                     \frac{1}{m}\sum_{i=1}^m y^{(i)} X_k^{(i)} \\
                     %
        \end{aligned}

Thinking a while about the structure of the operations involved, we can see that each sum corresponds to the inner (dot) product of two vectors formed with the observations of each variable. If we agree to represent (or collect) the observations of each variable in a matrix of the form:

\begin{aligned}
       X = 
        \begin{pmatrix}
               X_0^{(1)} \quad X_1^{(1)} \quad X_2^{(1)} \quad \cdots \quad X_n^{(1)} \\
               X_0^{(2)} \quad X_1^{(2)} \quad X_2^{(2)} \quad \cdots \quad X_n^{(2)} \\
               X_0^{(3)} \quad X_1^{(3)} \quad X_2^{(3)} \quad \cdots \quad X_n^{(3)} \\
                \vdots   \quad  \vdots   \quad \vdots    \quad \vdots \quad \vdots  \\
               X_0^{(m)} \quad X_1^{(m)} \quad X_2^{(m)} \quad \cdots \quad X_n^{(m)}
         \end{pmatrix}
\end{aligned}

Using this matrix representation for $X$, the derivatives of $J(\theta)$ can be written in the form:

\begin{aligned} 
        \begin{pmatrix} 
             \frac{\partial J(\theta)}{\!\!\!\! \partial \theta_0} \\ 
             \frac{\partial J(\theta)}{\!\!\!\! \partial \theta_1} \\
             \frac{\partial J(\theta)}{\!\!\!\! \partial \theta_2} \\ 
                        \vdots \\
             \frac{\partial J(\theta)}{\!\!\!\! \partial \theta_n} 
         \end{pmatrix} = \frac{1}{m}
%%%%%%%%%%%%%%
         \begin{pmatrix}
               X_0^{(1)} \quad X_0^{(2)} \quad X_0^{(3)} \quad \cdots \quad X_n^{(m)} \\
               X_1^{(1)} \quad X_1^{(2)} \quad X_1^{(3)} \quad \cdots \quad X_n^{(m)} \\
               X_2^{(1)} \quad X_2^{(2)} \quad X_2^{(3)} \quad \cdots \quad X_n^{(m)} \\
                \vdots   \quad  \vdots   \quad \vdots    \quad \vdots \quad \vdots  \\
               X_n^{(1)} \quad X_n^{(2)} \quad X_n^{(3)} \quad \cdots \quad X_n^{(m)}
         \end{pmatrix}
%%%%%%%%%%%%%%
        \begin{pmatrix}
               X_0^{(1)} \quad X_1^{(1)} \quad X_2^{(1)} \quad \cdots \quad X_n^{(1)} \\
               X_0^{(2)} \quad X_1^{(2)} \quad X_2^{(2)} \quad \cdots \quad X_n^{(2)} \\
               X_0^{(3)} \quad X_1^{(3)} \quad X_2^{(3)} \quad \cdots \quad X_n^{(3)} \\
                \vdots   \quad  \vdots   \quad \vdots    \quad \vdots \quad \vdots  \\
               X_0^{(m)} \quad X_1^{(m)} \quad X_2^{(m)} \quad \cdots \quad X_n^{(m)}
        \end{pmatrix}
%%%%%%%%%%%%%%
        \begin{pmatrix} \theta_0 \\ 
                        \theta_1 \\
                        \theta_2 \\ 
                        \vdots \\
                        \theta_n 
         \end{pmatrix}
- \frac{1}{m}
         \begin{pmatrix}
               X_0^{(1)} \quad X_0^{(2)} \quad X_0^{(3)} \quad \cdots \quad X_n^{(m)} \\
               X_1^{(1)} \quad X_1^{(2)} \quad X_1^{(3)} \quad \cdots \quad X_n^{(m)} \\
               X_2^{(1)} \quad X_2^{(2)} \quad X_2^{(3)} \quad \cdots \quad X_n^{(m)} \\

               X_n^{(1)} \quad X_n^{(2)} \quad X_n^{(3)} \quad \cdots \quad X_n^{(m)}
         \end{pmatrix}
        \begin{pmatrix} y^{(1)} \\ 
                        y^{(2)} \\
                         \\ 
                        \vdots \\
                        y^{(m)}
         \end{pmatrix}
\end{aligned}
